```python
import sys
import os,glob,csv
import re, numpy as np, pandas as pd
from pprint import pprint
```

```python
candid_list = list()
for root, dirs, files in os.walk("/Users/hellofutrue/Insta", topdown=False):
    for name in dirs:
        candid_list.append(os.path.join(root, name))

candid_name = []
for i in range(0, len(candid_list)):
    name = os.path.basename(os.path.normpath(candid_list[i]))
    candid_name.append(name)   
```

```python
top_list = list()
for root, dirs, files in os.walk("/Volumes/Untitled/top", topdown=False):
    for name in dirs:
        top_list.append(os.path.join(root, name))

top_name = []
for i in range(0, len(top_list)):
    name = os.path.basename(os.path.normpath(top_list[i]))
    top_name.append(name)
```

```python
folder_list = candid_list + top_list
folder_name = candid_name + top_name
```

## summarize all posts from all influencers in a single list (each entry is one post)
- dat = list of single post regardless of influencers
- dat_join = list of posts, each one for each influencer

```python
#specify which folder to extract info.
listname = folder_list
```

```python
dat = []
dat_join = []
for path in listname:
    os.chdir(path)
    post = []
    for file in glob.glob("*.txt"): 
        for line in open(file, 'r'):
            post.append(line.strip())
            dp = ''.join(post)
    dat.extend(post)
    dat_join.append(post)
```

## Save out the read-in documents

```python
#all posts
posts_all = pd.DataFrame(np.array(dat).reshape(1,len(dat))).T
posts_all.to_csv('/Users/hellofutrue/Desktop/Insight/Python/Feb/files/posts_all.csv') 
```
